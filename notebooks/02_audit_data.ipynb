```python
import pandas as pd
from pathlib import Path

processed = Path("../data/processed")

results = pd.read_csv(processed / "eci_2024_candidate_results.csv")

results["party"].nunique()
```

```python
party_counts = (
    results["party"]
    .value_counts()
    .reset_index()
)

party_counts.columns = ["party", "candidate_count"]

party_counts.head(100)
```

```python
party_counts.to_csv(processed / "party_counts_audit.csv", index=False)
```

```python
party_meta = party_counts.copy()

party_meta["party_type"] = "unknown"
party_meta["alliance_2024"] = "OTHER"
party_meta["ideology"] = "unknown"
party_meta["region"] = "national"
```
